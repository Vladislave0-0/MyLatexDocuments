# Уравнения Максвелла. Волновое уравнение

\begin{equation*}
\begin{cases}
    div (\vec{D}) = 4\pi\rho_{св} = 0,
    \\
    div (\vec{B}) = 0,
    \\
    rot (\vec{E}) = -\frac{1}{c} \frac{\partial{\vec{B}}}{\partial{t}},
    \\
    rot (\vec{H}) = \frac{4\pi}{c}\overrightarrow{j_{св}}+\frac{1}{c}\frac{\partial{\vec{D}}}{\partial{t}}
\end{cases}
\end{equation*}

В оптике нет свободных зарядов, поэтому $\rho_{св} = 0$ и $\overrightarrow{j_{св}} = \vec{0}$

Также из прошлого семестра нам известно, что $\vec{B} = \mu\vec{H}$; $\vec{D} = \varepsilon\vec{E}$

Из (3): $rot(rot(\vec{E})) = -\frac{1}{c}\frac{\partial{}}{\partial{t}}(rot(\vec{B})) = grad(div(\vec{E})) - \nabla^{2}\vec{E}$

$div(\vec{E}) = \frac{1}{\varepsilon}div(\vec{D}) = \frac{4\pi}{\varepsilon}\rho_{out} = 0$

$-\nabla^{2}\vec{E} = -\frac{\mu}{c}\frac{\partial{}}{\partial{t}}rot(\vec{H}) = -\frac{\mu}{c}\frac{\partial{}}{\partial{t}}(\frac{4\pi}{c}\overrightarrow{j_{out}}+ \frac{1}{c}\frac{\partial{\vec{D}}}{\partial{t}})$,   $\overrightarrow{j_{out}} = \vec{0}$

тогда $\nabla^{2}\vec{E} - \frac{\varepsilon\mu}{c} \cdot \frac{\partial^{2}{\vec{E}}}{\partial{t^{2}}} = 0$ — волновое уравнение.

# Принцип Гюйгенса-Френеля

Каждый элемент волнового фронта является центром возмущения, порождающего вторичные сферические волны, в результате интерференции которых формируется результирующее световое поле в каждой точке пространства.

# Количественная формулировка принципа Гюйгенса-Френеля

$\sum$ — поверхность, в каждой точке которой известны $\vec{E}$ и $\varphi$.

$E_{p} = \int\limits_{\sum}E(\vec{r})\frac{e^{ix\rho}}{\rho}k(\varphi)d\sigma$

$\lim\limits_{\varphi\to 0} k(\varphi) = \frac{1}{i\lambda}$

# Общая задача о дифракции на тонком экране

Пусть $\widehat{E}_{s}(x, y)$ — волновое поле в плоскости $z = 0$, проходящее через экран.

Постановка задачи:

1) найти волновое поле $\widehat{E}_{0}(x, y)$ в плоскости $z = 0$

2) найти волновое поле $\widehat{E}(x, y, z)$ в области $z > 0$

Решить уравнение Гельмгольца с граничными условиями

\begin{equation*}
\begin{cases}
    \nabla^{2}E+\frac{\omega^2}{c^{2}}E = 0,
    \\
    \widehat{E}(x, y, 0) = \widehat{E}_{0}(x,y)
\end{cases}
\end{equation*}

# Дифракция Френеля на круглом отверстии

$$
E(z) = \int\limits_0^{2\pi}\int\limits_0^R E_0K(\varphi)\frac{e^{ik\rho}}{\rho}rdrd\varphi = 2\pi E_0 \int\limits_0^R K(\varphi)\frac{e^{ik\rho}}{\rho}rdr
$$

Считаем, что $\varphi \ll 1 \Rightarrow K(\varphi) \longrightarrow \frac{1}{i\lambda}$ при $\varphi \longrightarrow 0$, тогда:

$$
E(z) = \frac{2\pi E_0}{i\lambda} \int\limits_0^R \frac{e^{ik\rho}}{\rho}rdr
$$

Имеем $\rho^2 = z^2 + r^2 \Rightarrow 2 rdr = 2 \rho d\rho$

$$
E(z) = \frac{2\pi  E_0}{i\lambda} \int\limits_z^{\rho^*} e^{ik\rho}\rho = \frac{2\pi E_0}{i\lambda}\cdot\frac{e^{ik\rho}}{ik}\bigg|_z^{\rho^*} = E_0\left(e^{ik\rho} - e^{ik\rho^*}\right)
$$

$\rho^* = \sqrt{z^2+R^2} \approx z + \frac{R^2}{2z} = z + \Delta_{max} \Rightarrow E(z) = E_0e^{ikz}\left(1-e^{ik\Delta_{max}}\right)$


$$
I(z) = E_0^2\left(1-e^{ik\Delta_{max}}\right)\left(1-e^{-ik\Delta_{max}}\right) = E_0^2(2-2cos(k\Delta_{max})) = 4E_0^2\sin^2(\frac{k\Delta_{max}}{2}) = 4E_0^2\sin^2(\frac{kR^2}{4z})
$$

Итого: 
$$\boxed{
I(z) = 4E_0^2\sin^2(\frac{kR^2}{4z})}
$$

$\frac{kR^2}{4z_1} = \frac{\pi}{2} \Rightarrow z_{\text{Д}} = z_1 = \frac{R^2}{\lambda}$

$\frac{kR^2}{4z_s} = \frac{2\pi}{\lambda}\cdot\frac{R^2}{4z_s} = \frac{\pi}{2}s \Rightarrow z_s = \frac{R^2}{\lambda}\cdot\frac{1}{s}$ — запишем это по-разному:

$z_s = \frac{R^2}{\lambda s}; R_s = \sqrt{\lambda zs}$ — радиус $s$-той зоны Френеля; $s = \frac{R^2}{\lambda s}$ — число зон Френеля.

# Спираль Френеля

Перезодя к разбиению волнового фронта на бесконечно малые площадки, получаем непрерывную кривую, называемую **спиралью Френеля**. Спираль постепенно закручивается к центру.

Для длин волн, характерных свету, закручивание очень медленное вследствие условий $R \gg \lambda$, $x \gg \lambda$.

Считаем вклад от 1 кольца $\implies$ получаем вклад $d\vec{E}$

Считаем вклад от 2 кольца $\implies$ получаем следующий вклад + поворот

И так далее...

Если $s = 1$, $\partial{\varphi} = \pi$ $\implies$ I зона — первая полуокружность и так далее.

Если открыто много зон Френеля $\implies$ радиус спирали $E_{0}$

# Пятно Пуассона

Если на пути излучения источника поставлена пластинка, а остальная часть открыта, то в области геометрической тени можем наблюдать светлое пятно Пуассона.

Условием наблюдения является неровность края $\le 1$ зоны Френеля, иначе в точку придут волны от фрагментов соседних зон, подавляющие интерференционный максимум.

Поскольку $\Delta R = R_{m+1} - R_{m} = \sqrt{\lambda b}(\sqrt{m+1}-\sqrt{m}) \approx \frac{1}{2}\sqrt{\frac{\lambda b}{m}}$

Учитывая, что $R = \sqrt{\lambda bm}$ $\implies$ $m = \frac{R^{2}}{\lambda b}$ $\implies$ $\Delta R = \frac{\lambda b}{2R}$ - неровность края диска.

# Дифракция Фраунгофера

$E(\xi, \eta) = \iint E(x, y)\frac{e^{-il\rho}}{\rho}\frac{1}{i\lambda}dxdy$ - интеграл Френеля

Запишем этот интеграл в зоне Фраунгофера

1) $\frac{1}{\rho} \approx \frac{1}{z}$;

2) $e^{-ik\rho} \not\approx e^{-ikz}$

$\rho^{2} = z^{2}+(x - \xi)^{2} + (y - \eta)^{2} = z^2+\xi^{2}+\eta^{2}-2x\xi-2y\eta+x^{2}+y^{2} \implies$

$\rho = (b^{2}-2x\xi-2y\eta+x^2+y^2)^{\frac{1}{2}} = b(1 - 2\frac{x\xi+y\eta}{b^{2}} + \frac{x^2+y^2}{b^2})^{\frac{1}{2}}$

$\rho k = \frac{2\pi}{\lambda}b(1 - \frac{x\xi+y\eta}{b^{2}} + \frac{x^2+y^2}{2b^2}) = \frac{2\pi}{\lambda}b - 2\pi\frac{x\xi+y\eta}{\lambda b}+ \pi\frac{x^2+y^2}{b}$, где $\pi\frac{x^2+y^2}{b} \approx 0$

$E(\xi, \eta) = \frac{1}{i\lambda z}e^{-i\frac{2\pi b}{\lambda}} \int \int E(x, y) exp(ik(\frac{x\xi}{b}+\frac{y\eta}{b}))dxdy$. Выбросим размерный множитель.

$E(\xi, \eta) = \iint\limits_{-\infty}^{+\infty} E(x, y)e^{ik(\frac{x\xi}{b}+\frac{y\eta}{b})}dxdy$, где $\frac{\xi}{b} = sin(\theta)$ и $\frac{\eta}{b} = sin(\psi)$ — направляющие углы

$E(\theta, \psi) = \iint E(x, y) exp(i(kxsin(\theta)+kysin(\psi)))dxdy$

В одномерном случае: $E(\theta) = \int\limits_{-\infty}^{+\infty} E(x)exp(ikxsin(\theta))dx$

Итого: $E(k_{x}) = \int\limits_{-\infty}^{+\infty} E(x)e^{ik_{x}x}dx$ — интеграл Фурье в зоне Фраунгофера - прообраз Фурье.

# Дифракция на одномерной щели

$E(\theta) = \int\limits_{-\infty}^{+\infty} E(x)exp(ikxsin(\theta))dx = \int\limits_{-\frac{D}{2}}^{+\frac{D}{2}} E(x)exp(ikxsin(\theta))dx = E_{0}\frac{D}{\frac{Dksin(\theta)}{2}}sin(\frac{Dksin(\theta)}{2})$

$E(\theta) = E_{0}Dsinc(\frac{kD}{2}sin(\theta))$, $I(\theta) = E_{0}^{2}D^{2}sinc^{2}(\frac{kD}{2}sin(\theta))$

# Дифракция Фраунгофера на решётке. Разрешённая способность, область дисперсии, угловая дисперсия решётки.

1) $E(\theta) = \int\limits_{-\infty}^{+\infty} E(x)e^{ikx\sin(\theta)}dx = \int\limits_{0}^{D} E(x)e^{ikx\sin(\theta)}dx = \sum\limits_{j=0}^{N-1}\int\limits_{jb}^{(j+1)b} E(x)e^{ikx\sin(\theta)}dx$

2) Рассмотрим конкретный номер j:

$\int\limits_{jb}^{(j+1)b} E(x)e^{ikx\sin(\theta)}dx$

3) Сделаем замену $\tilde{x} = x -jb$:

$\int\limits_{0}^{b} E(\tilde{x})e^{ik\tilde{x}\sin(\theta)}e^{ikjb\sin(\theta)}d\tilde{x} = e^{ikjb\sin{\theta}} \int\limits_{0}^{b} E(\tilde{x})e^{ik\tilde{x}\sin(\theta)}d\tilde{x} = e^{ikjb\sin{\theta}} \cdot Int$

4) $E(\theta) = \sum\limits_{j=0}^{N-1}e^{ikjb\sin(\theta)} Int = Int \cdot \sum\limits_{j=0}^{N-1}e^{ikjb\sin(\theta)} = Int\cdot e^{i\zeta}\frac{\sin(\frac{1}{2}Nkb\sin(\theta))}{\sin(\frac{1}{2}kb\sin(\theta))}$

5) $Int = \int\limits_{0}^{a} E_0e^{ik\tilde{x}\sin(\theta)}d\tilde{x} = E_0\frac{e^{ik\tilde{x}\sin(\theta)}}{ik\sin(\theta)}\bigg|_0^a = 2E_0\frac{e^{ik\tilde{x}\sin(\theta)} - 1}{2ik\sin(\theta)} = aE_0e^{i\zeta}\frac{\sin(\frac{1}{2}ka\sin(\theta))}{\sin(\frac{1}{2}ka\sin(\theta))}$

6) Тогда получаем:


$$
\boxed{
E(\theta) = E_0a\cdot sinc(\frac{1}{2}ka\sin(\theta))\frac{\sin(\frac{1}{2}Nkb\sin(\theta))}{\sin(\frac{1}{2}kb\sin(\theta))}e^{i\psi}}
$$

$$
\boxed{
I(\theta) = E_0^2\cdot sinc^2(\frac{1}{2}ka\sin(\theta))\frac{\sin^2(\frac{1}{2}Nkb\sin(\theta))}{\sin^2(\frac{1}{2}kb\sin(\theta))}}
$$

Причём, $sinc^2(\frac{1}{2}ka\sin(\theta))$ отвечает за огибающую, $\sin^2(\frac{1}{2}Nkb\sin(\theta))$ — за ширину, $\sin^2(\frac{1}{2}kb\sin(\theta))$ — за положение.

Мы изобразили картину для одной длины волны. Для другой длины волны пики будут в других местах, причём расстояние между максимумами будет постоянно увеличиваться: 

$$
\frac{\lambda_2-\lambda_1}{b},\ 2\frac{\lambda_2-\lambda_1}{b}, \dotsc
$$

1) Разрешающая способность решётки: $m\frac{\delta\lambda}{b} \geqslant \frac{\lambda}{D} \Rightarrow \frac{\lambda}{\delta\lambda} \geqslant \frac{mD}{b} \Rightarrow  \boxed{R = \frac{\lambda}{\delta\lambda} = mN}$.

2) Если $n+1$-порядок одной волны начал пересекаться с $n$-порядком другой волны, это бы плохо повлияло на решётку как на спектральный прибор. Оценим дисперсионную область:

$$
m(\lambda + \Delta\lambda) = (m+1)\lambda \Rightarrow \boxed{\Delta\lambda = \frac{\lambda}{m}}.
$$

3) Оценим также угловую дисперсию решётки:

$$
b\sin(\theta_m) = m\lambda \Rightarrow b\cos(\theta_m)d\theta = md\lambda \Rightarrow \boxed{\frac{d\theta}{d\lambda} = \frac{m}{b\cos(\theta_m)}}.
$$

# Эксперимент 1: Дифракция на одиночной щели

Подключим необходимые библиотеки:

```python
import numpy as np
import scipy as sp
from scipy.fft import fft2
from scipy.fft import ifft2
from scipy.fft import fftfreq
from scipy.fft import fftshift
import imageio
import cv2
import scienceplots

import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
from matplotlib import animation
from matplotlib.animation import PillowWriter
import pint

plt.style.use(['science', 'notebook'])
u = pint.UnitRegistry()
```

Установим параметры щели (ширину щели, длину волны):

```python
D = 0.1 * u.mm          # длина щели 0.1 мм
lam = 555 * u.nm        # длина волны 660 нм
```

Добавим цвет для дифракционной картины:

```python
def wavelength_to_rgb(wavelength, gamma=0.8):
    wavelength = float(wavelength)
    if wavelength >= 380 and wavelength <= 440:
        attenuation = 0.3 + 0.7 * (wavelength - 380) / (440 - 380)
        R = ((-(wavelength - 440) / (440 - 380)) * attenuation) ** gamma
        G = 0.0
        B = (1.0 * attenuation) ** gamma
    elif wavelength >= 440 and wavelength <= 490:
        R = 0.0
        G = ((wavelength - 440) / (490 - 440)) ** gamma
        B = 1.0
    elif wavelength >= 490 and wavelength <= 510:
        R = 0.0
        G = 1.0
        B = (-(wavelength - 510) / (510 - 490)) ** gamma
    elif wavelength >= 510 and wavelength <= 580:
        R = ((wavelength - 510) / (580 - 510)) ** gamma
        G = 1.0
        B = 0.0
    elif wavelength >= 580 and wavelength <= 645:
        R = 1.0
        G = (-(wavelength - 645) / (645 - 580)) ** gamma
        B = 0.0
    elif wavelength >= 645 and wavelength <= 750:
        attenuation = 0.3 + 0.7 * (750 - wavelength) / (750 - 645)
        R = (1.0 * attenuation) ** gamma
        G = 0.0
        B = 0.0
    else:
        R = 0.5
        G = 0.7
        B = 0.4
    return (R, G, B)
```

```python
color_array= [lam]
cmaps = [LinearSegmentedColormap.from_list('custom', 
                                        [(0,0,0),wavelength_to_rgb(lam.magnitude)],
                                        N=256) for lam in color_array]
```

Определим пространственную сетку от -2 мм до 2 мм с шагом 1600 шагов (x и y).

```python
x = np.linspace(-2,2,1600) * u.mm 
xv, yv = np.meshgrid(x, x)              # область, в которой мы будем моделировать дифракционную картину
```

Определим $u(x,y,0)$; во всех примерах переменной будет `U0`. В данном случае мы хотим, чтобы это была прямоугольная щель.

```python
U0 = (np.abs(xv)< D/2) * (np.abs(yv)<0.5*u.mm)      # геометрическое место точек щели, по сути уравнения щели
U0 = U0.astype(float)
```

График `U0`:

```python
plt.figure(figsize=(10,10))
plt.pcolormesh(xv,yv,U0)
plt.show()
```

Теперь нам нужно получить $A(k_x, k_y)$, заданное первым уравнением в рамке выше. Кроме того, нам нужна сетка значений $k_x$ и $k_y$, чтобы мы могли в конечном итоге вычислить $e^{- i z \sqrt{k^2-k_x^2-k_y^2}}$ (который должен быть 2D-массив).

```python
A = fft2(U0)
kx = fftfreq(len(x), np.diff(x)[0]) * 2 * np.pi # домножаем на 2pi, чтобы получить угловую частоту
kxv, kyv = np.meshgrid(kx,kx)
```

Напишем функцию, которая возвращает $u(x,y,z)$. Это зависит от расстояния $z$ от щели и $k=2\pi/\lambda$ (длины волны света).

```python
def get_U(z,k):
    return ifft2(A*np.exp(1j*z*np.sqrt(k**2-kxv**2-kyv**2)))
```

Укажем расстояние от щели до экрана $z=d$ и $k$:

```python
k = 2*np.pi / (lam)
d = 3* u.cm
```

Получим значение $U$:

```python
U = get_U(d,k)
```

```python
figure1, figure2 = plt.subplots(1, 2, figsize=(20,10))
figure2[0].pcolormesh(xv,yv,U0)
figure2[1].pcolormesh(xv,yv,np.abs(U), cmap=cmaps[0])
plt.show()
```

Для однощелевого эксперимента минимумы интенсивности возникают при углах $\theta$ (щель и экран) таких, что

$$D\sin\theta = m\lambda$$

где $D$ $-$ это ширина щели и $m$ ненулевое целое число. Отмечая, что

$$\sin\theta = x/\sqrt{x^2 + d^2}$$

где $x$ $-$ горизонтальное расстояние на экране и $d$ $-$ расстояние от щели до экрана, мы можем найти позиции $x$ в минимумах:

$$x = \frac{m \lambda d}{\sqrt{D^2 - m^2 \lambda^2}}$$



Получим позиции минимумов:

```python
m  = np.arange(1,5,1)
x_min = np.sqrt(m**2 * lam**2 * d**2 / (D**2 - m**2 * lam**2)).to('mm')
```

```python
plt.plot(x, np.abs(U)[250]**2)
[plt.axvline(x.magnitude, ls = '--', color = 'r') for x in x_min]
[plt.axvline(-x.magnitude, ls = '--', color = 'r') for x in x_min]
plt.xlabel('$x$ [мм]')
plt.ylabel('$I(x)$ [интенсивность]')
plt.show()
```

# Эксперимент 2: Дифракция на двух щелях

Для последующих экспериментов напишем функцию, которая делает все быстрее:

```python
def compute_U(U0, xv, yv, lam, z):
    A = fft2(U0)
    kx = 2*np.pi * fftfreq(len(x), np.diff(x)[0])
    kxv, kyv = np.meshgrid(kx,kx)
    k = 2*np.pi/lam
    return ifft2(A*np.exp(1j*z*np.sqrt(k**2-kxv**2-kyv**2)))
```

Зададим расстояние между щелями $s$, ширину щели $D$, и используем её для получения начального поля $u(x,y,0)$

```python
S = 0.2*u.mm
D = 0.05*u.mm
x = np.linspace(-4,4,3200) * u.mm
xv, yv = np.meshgrid(x, x)

U0 = (np.abs(xv-S/2)< D/2) * (np.abs(yv)<2*u.mm) + (np.abs(xv+S/2)< D/2) * (np.abs(yv)<2*u.mm)
U0 = U0.astype(float)
```

Покажем начальное поле:

```python
plt.figure(figsize=(10,10))
plt.pcolormesh(xv,yv,U0)
plt.show()
```

Вычислим $u(x,y,z)$ при $z=5$ см

```python
U = compute_U(U0, xv, yv, lam, z=5*u.cm)
```

Построим дифракционную картину

```python
figure1, figure2 = plt.subplots(1, 2, figsize=(20,10))
figure2[0].pcolormesh(xv,yv,U0)
figure2[1].pcolormesh(xv,yv,np.abs(U), cmap=cmaps[0])
plt.show()
```

Построим график напряжённости в зависимости от координаты на экране:

```python
plt.plot(x, np.abs(U)[250]**2)
plt.xlabel('$x$ [мм]')
plt.ylabel('$I(x)$ [Интенсивность]')
plt.grid()
```
